# Problem 1 :
### Problem Statement :
Print no. of matches won by each team based on year wise (each year as one column)

# Problem 2 :
### Problem Statement :
Print ManOfMach count of each player in Hyderabad stadium

```python
import pandas,numpy
df=pandas.read_csv('DataFiles/Match.csv')
df2=pandas.read_csv('DataFiles/Match.csv',usecols=['Team1'])
teams=numpy.unique(df2.values)
players=[]
for i in range(len(df)):
    if df.values[i][14] not in teams:
        if df.values[i][6]=='Rajiv Gandhi International Stadium, Uppal':
            players.append(df.values[i][14])
player=numpy.unique(players)
number=[]
for x in player:
    count=0
    for i in range(len(players)):
        if x==players[i]:
            count=count+1
    number.append(count)
df2=pandas.DataFrame()
df2['players']=player
df2['number of ManOfMatch']=number
df2
```
