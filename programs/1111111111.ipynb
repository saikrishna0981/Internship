# Project

### Problem Statement
Based on the given dataset predict the House Sale Prices using  Linear  Regression.

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading Data From CSV File

```python
df1= pd.read_csv('DataFiles/AmesHousing.csv')
a,b,c,d,e,f,g,h=[],[],[],[],[],[],[],[]
for i in df1.values:
    a.append(i[5])
    b.append(i[81])
    c.append(i[47])
    d.append(i[67])
    e.append(i[63])
    f.append(i[44])
    g.append(i[45])
    h.append(i[39])
df=pd.DataFrame({'LotArea':a,'SalePrice':b,'GrLvArea':c,'WoodDeckSF':d,'GarageArea':e,'frstFlrSF':f,'scndFlrSF':g,'TotalBsmtSF':h})
df
```

### Missing values

```python
from sklearn.preprocessing import scale
a=df.fillna(method="ffill")
a
```

### Correlation

```python
a.corr()
```

### Pair Plotting

```python
sns.pairplot(a.iloc[:,:])
```

### R-squared value

```python
import statsmodels.formula.api as smf
```

```python
ml1 = smf.ols("SalePrice ~ LotArea+GrLvArea+WoodDeckSF+GarageArea+frstFlrSF+scndFlrSF+TotalBsmtSF", data = a).fit()
```

```python
ml1.summary()
```

### Parameters

```python
ml1.params
```

### Input & output values

```python
X= a.iloc[:,:1]
X
```

```python
y = a.iloc[:,1]
y
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state = 5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Prediction

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print(regressor.intercept_)
regressor.coef_
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

### Comparision

```python
y_pred == y_test
```

### Plotting

```python
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, regressor.predict(X_train))
plt.title('LotArea vs SalePrices')
plt.xlabel("LotArea")
plt.ylabel("SalePrices")
plt.show()
```

```python
plt.plot(X_train, regressor.predict(X_train))
plt.scatter(X_test, y_test, color = "cyan")
```

```python
plt.plot(X_test, regressor.predict(X_test))
plt.scatter(X_test, y_test, color = "cyan")
```
